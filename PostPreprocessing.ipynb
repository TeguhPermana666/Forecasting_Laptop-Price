### Encoding categorical to numerical 

```python
def encoding(features):
    df[features + '_name_encoded'] = LabelEncoder().fit_transform(df[features])
```

```python
features = ["processor_name","OS","disk_drive","touchscreen","company","graphic_card"]
for i in features:
    encoding(i)
```
