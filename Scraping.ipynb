```python
from bs4 import BeautifulSoup 
import requests
import csv
import pandas as pd
import numpy as np
```

```python
link = ("https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page=1")
req = requests.get(link) # Requesting the content of the URL
content = req.content # getting content
```

# Html Parser
- Here we need to specify the content variable
- the parser which is the HTML parser
- so now soup is a variable of the beautiful soup object of result parsed HTML  

```python
soup = BeautifulSoup(content,'html.parser')
```

### Displays the whole code of website


```python
print(soup.prettify())
```

# Extraction the description from the website
- Using find method 
    - Grabbing the div tag which has the classname_4rR01T
-  So now this returns all the div tags with the classname of _4rR01T
- even can access the child tags with dot access.
- As class is a special keyword in python we have to use the class_ keyword
- pass the arguments here

```python
desc = soup.find_all('div' , class_='_4rR01T')
descriptions = []
for i in range(len(desc)):
    descriptions.append(desc[i].text) 
descriptions
len(descriptions)
```

# Extraction the spectification
- We observe that the classnames for the different specifications are under one div.So we need to apply some method to extract the different features.
- Extracting the text from the tags
- If RAM is present in the text then append it to the ram list. Similarly do this for the other features as well

```python
commonclass = soup.find_all('li',class_='rgWa7D')
# Create empty lists for the features
processors=[]
ram=[]
os=[]
storage=[]
inches=[]
warranty=[]

for i in range(0,len(commonclass)):
    p=commonclass[i].text 
    if("Core" in p): 
        processors.append(p)
    elif("RAM" in p): 
        ram.append(p)
    elif("HDD" in p or "SSD" in p):
        storage.append(p)
    elif("Operating" in p):
        os.append(p)
    elif("Display" in p):
        inches.append(p)
    elif("Warranty" in p):
        warranty.append(p)
```

# All Page

```python
descriptions = []
# Create empty lists for the features
processors=[]
ram=[]
os=[]
storage=[]
inches=[]
warranty=[]
prices = []
for i in range(1,65):
    link = ("https://www.flipkart.com/search?q=laptops&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={}".format(i))
    req = requests.get(link) # Requesting the content of the URL
    content = req.content # getting content
    soup = BeautifulSoup(content,'html.parser')
    desc = soup.find_all('div' , class_='_4rR01T')
    for i in range(len(desc)):
        descriptions.append(desc[i].text)
    
    commonclass = soup.find_all('li',class_='rgWa7D')
    for i in range(0,len(commonclass)):
        p=commonclass[i].text 
        if("Core" in p): 
            processors.append(p)
        elif("RAM" in p): 
            ram.append(p)
        elif("HDD" in p or "SSD" in p):
            storage.append(p)
        elif("Operating" in p):
            os.append(p)
        elif("Display" in p):
            inches.append(p)
        elif("Warranty" in p):
            warranty.append(p)
    price = soup.find_all('div',class_='_30jeq3 _1_WHN1') 
    for i in range(len(price)):
        prices.append(price[i].text)

d = {'Description':descriptions,'Processor':processors,'RAM':ram,'Operating System':os,'Storage':storage,'Display':inches,'Warranty':warranty,'Price':prices}
full_dataset = pd.DataFrame.from_dict(data = d,orient='index')
full_dataset = full_dataset.transpose() 
    
```
